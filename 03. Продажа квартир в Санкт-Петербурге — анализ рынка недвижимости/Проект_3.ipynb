# Исследование объявлений о продаже квартир
В вашем распоряжении данные сервиса Яндекc Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости.  
Ваша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма.  
# Описание данных
- airports_nearest — расстояние до ближайшего аэропорта в метрах (м)
- balcony — число балконов
- ceiling_height — высота потолков (м)
- cityCenters_nearest — расстояние до центра города (м)
- days_exposition — сколько дней было размещено объявление (от публикации до снятия)
- first_day_exposition — дата публикации
- floor — этаж
- floors_total — всего этажей в доме
- is_apartment — апартаменты (булев тип)
- kitchen_area — площадь кухни в квадратных метрах (м²)
- last_price — цена на момент снятия с публикации
- living_area — жилая площадь в квадратных метрах (м²)
- locality_name — название населённого пункта
- open_plan — свободная планировка (булев тип)
- parks_around3000 — число парков в радиусе 3 км
- parks_nearest — расстояние до ближайшего парка (м)
- ponds_around3000 — число водоёмов в радиусе 3 км
- ponds_nearest — расстояние до ближайшего водоёма (м)
- rooms — число комнат
- studio — квартира-студия (булев тип)
- total_area — площадь квартиры в квадратных метрах (м²)
- total_images — число фотографий квартиры в объявлении

К полученным данным, представленным на гистограмме, применимо определение «распределение Пуассона». На рынке превалируют квартиры с площадью, приближенной к 30 кВ.м. В силу того, что данное значение по СП 54.13330.2016. Является наименьшим рекомендуемым числом, применимым к определении площади квартир. Рекомендуемый диапазон площадей 1-комн кВ. Находится в пределах от 28 до 38. Жилая комната в 1-к квартире от 14 кв.м.,кухня от 5 кв.м. На гистограммах жилой площади и кухни так же видно скачок на минимально рекомендуемой площади.

```python
def without_hue(ax, feature): # добавим функция для подсчета процентов
    total = len(feature)
    for p in ax.patches:
        percentage = '{:.2f}%'.format(100 * p.get_height()/total)
        x = p.get_x() + p.get_width() / 2 - 0.05
        y = p.get_y() + p.get_height()
        ax.annotate(percentage, (x, y), size = 12)
```

```python
plt.figure(figsize=(9, 6))
ax = sns.countplot(x='rooms', palette='hls', data=data)
plt.xlim(-1,9)
plt.title('Количество квартир по комнатности')
plt.xlabel('Количество комнат')
plt.ylabel('Частота')
without_hue(ax,data['rooms'])
plt.show()
```

```python
plt.figure(figsize=(16, 5))
plt.xlim(0,20)
sns.boxplot(x=data['rooms'],color='green');
plt.title('Количество комнат в квартире')
plt.ylabel('rooms')
plt.xlabel('Количество предложений')
plt.show()

print('Среднее количество:', round(data['rooms'].mean(), 2))
print('Медиана:', round(data['rooms'].median(), 2))
```

В данных о высоте потолков было большое количество пропусков,
которое заполнено медианой, на гистограмме это значение хорошо выражено. 

```python
def stat_report(column, label,lim,bins):
    X_1 = column
    
    plt.figure(figsize =(20, 6))
    plt.xlim(0,lim)
    hist = X_1.plot(label= 'Значение',
                    kind='hist', 
                    bins=bins, 
                    alpha=0.5, 
                    ec='black',
                    color='green')
    q1 = column.quantile(0.25)                 
    q3 = column.quantile(0.75)
    iqr = q3 - q1
    plt.axvline(label= 'Среднее значение',x=X_1.mean(),linewidth=3, color='y', linestyle='--')
    plt.axvline(label= 'Медиана',x=X_1.median(),linewidth=3, color='orange', linestyle='--')    
    plt.axvline(label= 'Нижняя граница',x=q1 -(1.5*iqr),
                    linewidth=3, color='red',linestyle='--')
    plt.axvline(label= 'Верхняя граница', x=q3 +(1.5*iqr),
                    linewidth=3, color='b',linestyle='--')
    plt.title(label)
    plt.legend()
    plt.xlabel('Значение')
    plt.ylabel('Частота')
    plt.grid()
    plt.show()
    #Выведем на экран основные паказатели
    print('Среднее количество:', round(column.mean(), 2))
    print('Медиана:', round(column.median(), 2))
    print('Стандартное отклонение:', round(np.sqrt(np.var(column)), 2))
    print('Верхняя граница',q3 +(1.5*iqr))
```

```python
stat_report(data['last_price'],'0',30000000,1000)
```

Диаграмма размаха,показывает что медиана находится ближе к 4.65млн,а выбросы если стоимость квартиры превышает 12 млн.

```python
stat_report(data['price_square_metr'],'0',600000,1000)
```

Гистограмма имеет нормальное распределение. Пик со стоимостью квадратного метра находится на 100 000р.Выбросы при значениях больше 171 000.
